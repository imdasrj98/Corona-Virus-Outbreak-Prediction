```python
import pandas as pd
```

```python
country=pd.read_csv('covid_confirmed_daily_updates.csv')
```

```python
confirmed=pd.read_csv('confirm_predict.csv')
deaths=pd.read_csv('death_predict.csv')
recovered=pd.read_csv('recover_predict.csv')
```

```python
confirmed=confirmed[['0']]
deaths=deaths[['0']]
recovered=recovered[['0']]
```

```python
confirmed.columns=[['Confirmed']]
deaths.columns=[['Death']]
recovered.columns=[['Recovered']]
```

```python
country=country[['Country/Region']]
```

```python
result=pd.concat([country, confirmed, recovered, deaths], axis=1, sort=False)
```

```python
result.columns=[['Country/Region', 'Confirmed', 'Recovered', 'Death']]
```
